```python
import numpy as np
import pandas as pd
```

## Data Preparation

#### to transform pdf to csv https://www.zamzar.com/convert/pdf-to-csv/

```python
data = pd.read_csv("data.csv")
data.head(n=60)
```

```python
data=data.drop("(All)",axis=1)
data=data.drop("(All).1",axis=1)
data=data.drop("Unnamed: 15",axis=1)
data
```

```python
prepr = data.to_numpy()
```

```python
for k in range(13):
    i=0
    if k not in [1,10,11,12]:
        for j in prepr[:,k]:
            if np.isnan(j):
                prepr[:,k][i] = prepr[:,k][i-1]

            i = i + 1
            
    if k == 1:
        for i in range(len(prepr[:,1])):
            try:
                x=len(prepr[:,k][i])
            except:
                prepr[:,k][i] = prepr[:,k][i-1]
                
    if k == 10 or k == 11:
        for i in range(len(prepr[:,1])):
            try:
                x=len(prepr[:,k][i])
            except:
                prepr[:,k][i] = '0'
    
    if k == 12:
        for i in range(len(prepr[:,1])):
            if np.isnan(prepr[:,k][i]):
                prepr[:,k][i] = '0'
```

```python
data = pd.DataFrame(data=prepr,columns=["id_produit","catégorie","prix","weight_g","length_cm","height_cm","width_cm","mois","trimestre","annee","moyenne chiffre d'affaire","chiffre d'affaire","nombre de ventes"])
data
```

```python
data.to_csv(r'new_data.csv', index = True)
```

## Importing the dataset

```python
X  = np.array(data[['prix', 'weight_g', 'length_cm', 'height_cm','width_cm', 'mois', 'trimestre', 'annee']])
y  = np.array(data["nombre de ventes"]).astype('int')
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

## Random Forest (Regression)

```python
from sklearn.ensemble import RandomForestRegressor

clf1 = RandomForestRegressor(n_estimators = 1000, random_state = 42)
clf1.fit(X_train, y_train)


sum = 0
for x_p,y_p in zip(X_train, y_train):
    sum = sum + abs(clf1.predict([x_p])-y_p)
    
print("Erreur d'apprentissage:",(sum/np.sum(y_train)*100)[0], "%")

sum = 0
for x_p,y_p in zip(X_test,y_test):
    sum = sum + abs((clf1.predict([x_p])-y_p))
    
print("Erreur validation: ", (sum/np.sum(y_test)*100)[0], "%")
```

## SVR (Regression)

```python
from sklearn.svm import SVR

clf2 = SVR()
clf2.fit(X_train, y_train)


sum = 0
for x_p,y_p in zip(X_train, y_train):
    sum = sum + abs(clf2.predict([x_p])-y_p)
    
print("Erreur d'apprentissage:",(sum/np.sum(y_train)*100)[0], "%")

sum = 0
for x_p,y_p in zip(X_test,y_test):
    sum = sum + abs((clf2.predict([x_p])-y_p))
    
print("Erreur validation: ", (sum/np.sum(y_test)*100)[0], "%")
```

## KNN (Approchement)

```python
from sklearn.neighbors import KNeighborsClassifier

clf3 = KNeighborsClassifier(n_neighbors=1)
clf3.fit(X_train, y_train)


sum = 0
for x_p,y_p in zip(X_train, y_train):
    sum = sum + abs(clf3.predict([x_p])-y_p)
    
print("Erreur d'apprentissage:",(sum/np.sum(y_train)*100)[0], "%")

sum = 0
for x_p,y_p in zip(X_test,y_test):
    sum = sum + abs((clf3.predict([x_p])-y_p))
    
print("Erreur validation: ", (sum/np.sum(y_test)*100)[0], "%")
```
